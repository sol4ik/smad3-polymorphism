# Process FASTA files

```python
import os
```

```python
species = ["human", "mouse", "cow"]
```

```python
utr_start = {
    "human": 1495,
    "mouse": 1282,
    "cow": 1510
}
```

```python
DATA_DIR = "../data/"
```

```python
for specie in species:
    with open(os.path.join(DATA_DIR, "{}.fasta".format(specie)), "r") as fasta_file:
        lines = fasta_file.readlines()[1:]
    lines = ''.join(lines)
    utr3_segmet = lines[utr_start[specie]:]
    
    with open(os.path.join(DATA_DIR, "{}3'utr.fasta".format(specie)), "w") as fasta_file:
        fasta_file.write(utr3_segmet)
```
