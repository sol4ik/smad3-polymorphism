## Polymorphism in AREs search

```python
def parse_variation(variation):
    divider = variation.find('>')
    starting_allele = variation[:divider]
    polymorphed_alleles = tuple([ch for ch in variation[divider + 1:].strip().split('|')])
    return starting_allele, polymorphed_alleles
```

```python
are_with_snp = list()

no_pos = 0
no_allele = 0

for _, are in tqdm(ares.iterrows()):
    for _, snp in snps.iterrows(): 
        if are["Start"] <= int(snp["pos"]) <= are["End"]:
            pos = int(snp["pos"])
            if pos > 0:
                s_allel, poly_allels = parse_variation(snp["variation"])
                motif = are["Motif"]
                if s_allel != '-':
                    if motif[are["Start"] - pos:are["Start"] - pos + len(s_allel)] != s_allel:
                        print("! allele doesn't coinside with motif sequence")
                        no_allele += 1
                        continue
                alleles = '[' + s_allel
                for allele in poly_allels:
                    alleles += "/" + allele
                alleles += ']'
                motif = motif[:are["Start"] - pos] + alleles + motif[are["Start"] - pos + len(s_allel):]

                are_with_snp.append((motif, are["Start"], are["End"]))
                print("> successfully found polymorphism in ARE: {}".format(motif))
            else:
                print("! posisiton of given polymorphism is not specified")

print("\n\n")
print("successfully found polymorphism in ARE: {}/{}".format(len(ares) - no_pos - no_allele, len(ares)))
print("polymorphism position not provided: {}/{}".format(no_pos, len(ares)))
print("allele doesn't coinside with motif sequence: {}/{}".format(no_allele, len(ares)))
```
